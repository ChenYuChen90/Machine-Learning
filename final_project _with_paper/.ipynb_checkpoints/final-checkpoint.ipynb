```python
total = 441
for i in range(len(data.values)):
    for j in range(10):
        temp = {'png':'png_' + f'{total+j:03}',
                'name':data.values[i][1],
                'gender':data.values[i][2],
                'age':data.values[i][3],
                'case':data.values[i][4],
                'murderer':data.values[i][5]}
        data = data.append(temp, ignore_index=True)
    total += 10
    #print(data.values[i])
    #temp = {'png':'png_001','name':'hi', 'gender':'hi', 'age':0, 'case':2, 'murderer':5}
    #print(temp)
    
```

```python
PATH = "Test_data"  #path to your testing image
file_dir = os.listdir(PATH)
file_dir.sort()
len(file_dir)
```

```python
temp = 0
total = 441
check = 0
for i in range(len(file_dir)):
    if(int(i/10) - int((i-1)/10) > 0):
        temp = 0
    check %= 10
    if(check != temp):
        print(int(i/10))
        print(total)
        break
    if file_dir[i][:7] == 'png_' +  f'{int(i/10)+1:03}':
        temp_string = file_dir[i][:4] + str(total+temp) + file_dir[i][-4:]
        os.rename('Test_data/'+file_dir[i],'Test_data/'+temp_string)
        temp += 1
        check += 1
        if(temp % 10 == 0):
            total += 10
```
